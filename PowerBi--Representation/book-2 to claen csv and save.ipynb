```python
import pandas as pd
```

```python
data = pd.read_csv('API_SP.POP.TOTL.csv')
data
```

```python
data.rename(columns={
    data.columns[0]: "Country Name",
    data.columns[1]: "Country Code"
}, inplace=True)
data
```

```python
year_cols = [col for col in data.columns if col.isdigit()]
df = data[["Country Name", "Country Code"] + year_cols]
df
```

```python
df_unpivoted = df.melt(id_vars=["Country Name", "Country Code"],
                       var_name="Year",
                       value_name="Population")
df_unpivoted
```

```python
df_unpivoted["Year"] = df_unpivoted["Year"].astype(int)
df_unpivoted["Population"] = pd.to_numeric(df_unpivoted["Population"], errors="coerce")

df_unpivoted
```

```python
df_unpivoted.to_csv("CLEANED_TASK1.csv", index=False)
print("Cleaned file saved as CLEANED_TASK1.csv")
```
